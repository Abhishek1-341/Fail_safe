### using Mutual Information (MI)

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.preprocessing import LabelEncoder
```

```python
# Label encoding categorical columns
df_cat = df_maths.select_dtypes(include=['str']).copy()
for col in df_cat.columns:
    df_cat[col] = LabelEncoder().fit_transform(df_cat[col])
```

```python
# Calculate MI
mi_scores = mutual_info_regression(df_cat, df_maths['G3'], discrete_features=True)

mi_results = pd.Series(mi_scores, index=df_cat.columns).sort_values(ascending=False)
print(mi_results)
```

```python
df_cat = df_por.select_dtypes(include=['str']).copy()
for col in df_cat.columns:
    df_cat[col] = LabelEncoder().fit_transform(df_cat[col])

mi_scores = mutual_info_regression(df_cat, df_por['G3'], discrete_features=True)
mi_results = pd.Series(mi_scores, index=df_cat.columns).sort_values(ascending=False)
mi_results
```
